```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('covid19 lang:ja since:2020-06-01 until:2020-06-10').get_items()):
    if i>4000:
        break
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.lang])
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'lang'])
```

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

def getLanguageTweets(scrape_text):
    tweet_list=[]
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(scrape_text).get_items()):
        if i>400:
            break
        tweet_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.lang])

    # Creating a dataframe from the tweets list above
    tweets_data = pd.DataFrame(tweet_list, columns=['Datetime', 'Tweet Id', 'Text', 'Username', 'lang'])
    return tweets_data
df1= getLanguageTweets('covid19 lang:ja since:2020-06-01 until:2020-06-10')
```
